```python
def format_date(date_str):
    if not pd.isna(date_str):
        day, month, year = date_str.split('/')
        day = day.zfill(2)
        month = month.zfill(2)
        if len(year) == 2:
            year = '20' + year
        elif len(year) == 3:
            year = '2' + year
        return f'{day}/{month}/{year}'
    else:
        return date_str
```

```python
format_date(df['תאריך אסמכתא'].iloc[2])
```

```python
format_date(df['תאריך אסמכתא'].iloc[140])
```

```python
df['תאריך אסמכתא'] = df['תאריך אסמכתא'].apply(format_date)
```

```python
def safe_convert_dates(date_series, date_format):
    # Attempt to convert dates
    try:
        converted_dates = pd.to_datetime(date_series, format=date_format, errors='coerce')
        # Now using dashes instead of slashes
        return converted_dates.dt.strftime('%d-%m-%Y')
    except ValueError as e:
        print(f"Date conversion failed: {e}")
        return pd.Series([None] * len(date_series))

# Apply the function to the DataFrame
df['formatted_date'] = safe_convert_dates(df['תאריך אסמכתא'], '%d-%m-%Y')
```
